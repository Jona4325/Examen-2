# Errores
Calcule el eeror relativo para diferentes valores de N.

```python
import numpy as np
import matplotlib.pyplot as plt

# Definir la función de la ODE
def f(t, y):
    return -5 * y + 5 * t**2 + 2 * t

# Solución analítica
def y_analitica(t):
    return t**2 + (1/3) * np.exp(-5 * t)

# Método de Euler
def euler_method(f, t0, y0, h, N):
    t_values = [t0]
    y_values = [y0]
    t = t0
    y = y0
    for _ in range(N):
        y = y + h * f(t, y)
        t = t + h
        t_values.append(t)
        y_values.append(y)
    return np.array(t_values), np.array(y_values)

# Parámetros
t0 = 0
y0 = 1/3
a = 0
b = 1

# Calcular h
N = 10
h = (b - a) / N
print(f"Valor de h: {h}")

# Resolver la ODE usando el método de Euler
t_values, y_values = euler_method(f, t0, y0, h, N)

# Calcular la solución analítica
y_analitica_values = y_analitica(t_values)

# Calcular el error relativo
error_relativo = np.abs((y_analitica_values - y_values) / y_analitica_values)
promedio_error_relativo = np.mean(error_relativo)
print(f"Promedio del error relativo para N=10: {promedio_error_relativo:.2e}")

# Graficar las soluciones
plt.plot(t_values, y_values, label='Método de Euler')
plt.plot(t_values, y_analitica_values, label='Solución Analítica')
plt.xlabel('t')
plt.ylabel('y')
plt.legend()
plt.title('Solución de la ODE usando el método de Euler (N=10)')
plt.show()
```
